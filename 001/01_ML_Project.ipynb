```python
train.drop(['Cabin', 'Embarked', 'Fare', 'Ticket', 'Name', 'PassengerId'], axis=1, inplace=True)
```

```python
train.head()
```

```python
from math import isnan

def getAge(age, pclass):
  if not isnan(age):
    return age
  return int(train.groupby('Pclass')['Age'].mean()[pclass])
```

```python
train['Age'] = train.apply(lambda col: getAge(col.Age, col.Pclass), axis=1)
```

```python
train['Sex'].unique()
```

```python
train['Sex'] = train['Sex'].apply(lambda sex: 1 if sex.lower() == 'male' else 0)
```

```python
train.isnull().sum()
```

```python
train['Age'] = train['Age'].astype('int32')
```

```python
train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
train.head()
```

```python
train2 = pd.get_dummies(data=train, columns=['Pclass', 'SibSp', 'Parch'], dtype=int)
```

```python
train2.head()
```

```python
X,y = train.drop('Survived', axis=1), train['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=60)
```

```python
X_train.shape, X_test.shape
```

```python
rfc = RandomForestClassifier(n_estimators=120, criterion='gini')
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
sns.heatmap(data=confusion_matrix(y_test,y_pred), annot=True, fmt='d', cmap='Blues', xticklabels=['Not Survived', 'Survived'], yticklabels=['Not Survived', 'Survived'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix for Titanic Survival Prediction')
plt.show()
```
